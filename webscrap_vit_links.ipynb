```python
# Import libraries
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
#set the URL you want to webscrape from
url = 'http://chennai.vit.ac.in/academics/schools'
```

```python
#connect to the URL
response = requests.get(url)
```

```python
#all the script is taken into the responce varible
print(response.content) 
```

```python
# Parse HTML and save to BeautifulSoup object
soup = BeautifulSoup(response.content, 'lxml')
print(soup.prettify()) 
```

```python
hi = soup.select("li")
hi
```

```python
for i in range(len(hi)):
    print(hi[i].getText())
```

```python
for link in soup.find_all('a',href=True):
    print(link['href'])
```

```python
for link in soup.find_all('a',href=True):
    if(link['href'].startswith("/")):
        print("http://chennai.vit.ac.in/academics/schools"+link['href'])
    else:
        if((link['href'].startswith("http"))):
            print(link['href'])
```
