```python
import time
from selenium import webdriver
```

```python
driv_path=r"C:/Users/Tallam Tharun Sai/Downloads/chromedriver.exe"
driver=webdriver.Chrome(driv_path)
```

```python
driver.get("http://chennai.vit.ac.in/academics/schools/scse")
time.sleep(10)
btn=driver.find_element_by_xpath("""//*[@id="main"]/div/div/div[2]/div[2]/div/div/div[3]/a""")
time.sleep(10)
btn.click()
time.sleep(10)
print("\n School of Computing Science and Engineering - SCSE\n\n")
fac=driver.find_elements_by_xpath('//h6[@class="vitblue"]')
for i in range(len(fac)):
    print(fac[i].text)
```
